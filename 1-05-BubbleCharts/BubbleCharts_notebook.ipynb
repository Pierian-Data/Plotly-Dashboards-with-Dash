```python
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
```

```python
# load the data
df = pd.read_csv('C:/Users/20194136/Documents/Courses/Plotly-Dashboards-with-Dash/Data/mpg.csv')
print(df)
print(df.columns)

data = [go.Scatter(x=df['horsepower'],
                  y=df['mpg'],
                  text=df['name'],
                  mode='markers',
                  marker=dict(size=df['weight']/200, color=df['cylinders'], showscale=True))]

layout = go.Layout(title='horsepower, mpg and cylinders for different cars')

fig = go.Figure(data=data, layout=layout)

pyo.plot(fig)
```
